```python
from datetime import datetime

class Book:
    def __init__(self, title, author, year, quantity):
        self.title = title
        self.author = author
        self.year = year
        self.quantity = quantity
        self.time = 0  # tính thời gian mượn và trả sách

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year: {self.year}")
        print(f"Quantity: {self.quantity}")


class Library:
    def __init__(self):
        self.books = []   # lưu trữ tất cả sách trong thư viện

    def add_book(self, book):
        self.books.append(book)

    def display_all_books(self):
        if len(self.books) == 0:
            print("No books found.")
        else:
            print("Book List:")
            for book in self.books:
                book.display_info()
                print()

    def search_book(self, keyword):
        found_books = []
        for book in self.books:
            if keyword.lower() in book.title.lower() or keyword.lower() in book.author.lower():
                found_books.append(book)

        if len(found_books) == 0:
            print("No books found.")
        else:
            print("Found Books:")
            for book in found_books:
                book.display_info()
                print()

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.quantity > 0:
                    book.quantity -= 1
                    book.time = datetime.now()
                    print(f"You have borrowed '{book.title}', time: {book.time}.")
                else:
                    print(f"'{book.title}' is out of stock.")
                return

        print("Book not found.")

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.quantity += 1
                duration = datetime.now() - book.time
                
                print(f"You have returned '{book.title}'.")
                print(f"Duration of borrowing: {duration}.")
                return

        print("Book not found.")


def main():
    library = Library()

    while True:
        print("-" * 30)
        print("Library Management System")
        print("1. Add Book")
        print("2. Display All Books")
        print("3. Search Book")
        print("4. Borrow Book")
        print("5. Return Book")
        print("0. Exit")
        
        choice = input("Enter your choice: ")

        print("-" * 30)
        if choice == "1":
            title = input("Enter book title: ")
            author = input("Enter book author: ")
            year = input("Enter publication year: ")
            quantity = int(input("Enter quantity: "))

            book = Book(title, author, year, quantity)
            library.add_book(book)
            print("Book added successfully.")

        elif choice == "2":
            library.display_all_books()

        elif choice == "3":
            keyword = input("Enter book title or author: ")
            library.search_book(keyword)

        elif choice == "4":
            title = input("Enter book title to borrow: ")
            library.borrow_book(title)

        elif choice == "5":
            title = input("Enter book title to return: ")
            library.return_book(title)

        elif choice == "0":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
