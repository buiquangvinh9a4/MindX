```python
import matplotlib.pyplot as plt

x = list(df.loc[1:4,"Name"])   # chuẩn bị dữ liệu cho trục X
y = list(df.loc[1:4,"Area"])   # chuẩn bị dữ liệu cho trục Y

# cú pháp bar() để vẽ biểu đồ cột
plt.bar(x, y, width = 0.1,color="blue")
plt.title("Biểu đồ diện tích của một số tỉnh")
plt.xlabel("Tên tỉnh")
plt.ylabel("Diện tích (km2)")

plt.show()      # hiển thị biểu đồ

```

```python
import pandas as pd

df = pd.read_excel("provinces.xlsx")

x1 = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
y1 = [25.3, 26.7, 25.1, 20, 24.5, 27.8, 22.5, 29.9]


x2 = list(df.loc[1:4,"Name"])   # chuẩn bị dữ liệu cho trục X
y2 = list(df.loc[1:4,"Area"])   # chuẩn bị dữ liệu cho trục Y

x3 = [17, 56, 21, 56, 87]
color = ["blue", "yellow", "gray", "violet", "pink"]



# chia vùng biểu đồ
fig = plt.figure(figsize=(12,10))
(ax1, ax2), (ax3, ax4) = fig.subplots(2, 2)

ax1.plot(x1, y1)   # biểu đồ đường
ax1.set_title("Biểu đồ đường")

ax2.bar(x2, y2)    # biểu đồ cột
ax2.set_title("Biểu đồ cột")

ax3.plot(x2, y2, color="green")   # biểu đồ đường
ax3.set_title("Biểu đồ đường")

ax4.pie(x3, autopct='%1.1f%%', colors = color, startangle=90) # biểu đồ tròn
ax4.set_title("Biểu đồ tròn")
label = ["VietNam", "China", "Japan", "Korea", "Qatar"]
plt.legend(label, loc="upper right")

plt.show()
```

```python
# câu 1
df = pd.read_excel("pokemon_data.xlsx")
index = list(df["Type 1"].value_counts().index)
value = list(df["Type 1"].value_counts())

for i in range(len(index)):
    print(f"- {index[i]}: {value[i]}")
    
plt.pie(value, autopct='%1.1f%%')
plt.legend(index, loc="upper right")
plt.title("Biểu đồ % các hệ trong Pokemon")
plt.show()
```

```python
avg_hp = {}
for _type in index:
    avg = df.loc[df["Type 1"] == _type  , "HP"].mean()
    avg_hp[_type] = avg
    
avg_hp_sorted = dict(sorted(avg_hp.items(), key = lambda x : x[1], reverse = True))
avg_hp_sorted
```
