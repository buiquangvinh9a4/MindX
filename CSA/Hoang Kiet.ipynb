```python
class Book:
    def __init__(self, title, author, year, quantity):
        self.title = title
        self.author = author
        self.year = year
        self.quantity = quantity

    def display(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year of Publication: {self.year}")
        print(f"Quantity: {self.quantity}")

class Library(Book):



    def __init__(self):
        self.books = []
    def add_books(self,book):
        self.books.append(book)
        print(f'Added book {book.title}')
        
    def display_books(self):
        if not self.books:
            print("The library is empty")
        else:
            print("These are books in our library")
            for book in self.books:
                book.display()
    def search_book(self, keyword):
        found_books = [book for book in self.books if keyword.lower() in book.title.lower()]
        if not found_books:
            print(f"No books found with the keyword: {keyword}")
        else:
            print(f"Books found with the keyword '{keyword}':")
            for book in found_books:
                book.display()

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                if book.quantity > 0:
                    book.quantity -= 1
                    print(f"Borrowed book: {book.title}")
                else:
                    print(f"Sorry, '{book.title}' is out of stock.")
                return
        print(f"Book '{title}' is not available in the library.")

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                book.quantity += 1
                print(f"Returned book: {book.title}")
                return
        print(f"Book '{title}' is not part of the library's collection.")

library = Library()
book1 = Book("Bản lĩnh sói cô độc", "Hoang Kiet et.al.", 2023, 5)
book2 = Book("Shukusei! Loli kami requiem!", "Shigure Ui & IOSYS", 2023, 1)
book3 = Book("Đất rừng Phương Nam", "Đoàn Giỏi", 1957, 1)

library.add_books(book1)
library.add_books(book2)
library.add_books(book3)

library.display_books()

library.search_book("bản lĩnh")
library.search_book("loli kami")

library.borrow_book("Bản lĩnh sói cô độc")
library.borrow_book("Shukusei! Loli kami requiem!")
library.borrow_book("Đất rừng Phương Nam")

library.return_book("Bản lĩnh sói cô độc")
library.return_book("Đất rừng Phương Nam")



def menu():
    while True:
        print("\nLibrary Management System:")
        print("1. Add Book")
        print("2. Display All Books")
        print("3. Search Book")
        print("4. Borrow Book")
        print("5. Return Book")
        print("6. Exit")
        choice = input("Enter your choice by number from 1 to 6: ")
        if choice == '1':
            title = input('Enter book title you want to add:')
            author = input("Enter author: ")
            year = int(input("Enter year of publication: "))
            quantity = int(input("Enter quantity: "))
            book = Book(title, author, year, quantity)
            library.add_books(book)
        elif choice == '2':
            library.display_books()
        elif choice == '3':
            keyword = input("Enter the book you want to search: ")
            library.search_book(keyword)
        elif choice == '4':
            title = input("Enter the book to borrow: ")
            library.borrow_book(title)
        elif choice == '5':
            title = input("Book's title to return")
            library.return_book(title)
        elif choice == '6':
            print("Thanks for using our service!")
            break
        else:
            print("Invalid choice. Please select another option.")
menu()

```
