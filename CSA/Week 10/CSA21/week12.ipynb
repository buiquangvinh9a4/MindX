```python
import matplotlib.pyplot as plt   # thư viện vẽ biểu đồ
import pandas as pd    # thư viện xử lý dữ liệu

Ox = ["Tháng 1", "Tháng 2", "Tháng 3", "Tháng 4"]

Oy = [29.4, 27.6, 31.6, 20.7]

"""
    Các dạng biểu đồ:
        plt.bar() : biểu đồ cột 
        plt.scatter() : biểu đồ điểm
        plt.plot() : biểu đồ đường
        plt.pie()  : biểu đồ tròn
"""
plt.plot(Ox, Oy, color = "red")
plt.bar(Ox, Oy, color = "orange", width = 0.2)
plt.scatter(Ox, Oy, color = "green")
plt.title("Nhiệt độ trung bình của các tháng")
plt.xlabel("Tháng")
plt.ylabel("Nhiệt độ (C)")


plt.show()  # Hiển thị biểu đồ
```

```python
plt.pie(Oy,autopct='%1.1f%%', labels = Ox)
plt.title("Tỉ lệ nhiệt độ các tháng", fontsize = 20)
plt.show()
```

```python
# yêu cầu: đếm xem có bao nhiêu cầu thủ thuộc mức Overall sau:
df = pd.read_excel("trending_football_players.xlsx")
df
"""
    Mức 1: Dưới 80
    Mức 2: Từ 80 - 90
    Mức 3: Trên 90
"""

"""
    Thống kê độ tuổi của các cầu thủ thuộc có quốc tịch France
    - Dưới 20
    - Từ 20 đến 30
    - Từ 30 đến 40
    - Từ 40 trở lên
"""
xA = ["Dưới 80", "Từ 80 đến dưới 90", "Từ 90 trở lên"]
count_overall = [0, 0, 0]
count_overall[0] = int(df.loc[ (df["National_team"] == "Brazil") 
                    & (df["Overall"] < 80), ["Overall"]].count())
count_overall[1] = int(df.loc[ (df["National_team"] == "Brazil")
                    & ((df["Overall"] < 90) & (df["Overall"] >= 80)), ["Overall"]].count())
count_overall[2] = int(df.loc[ (df["National_team"] == "Brazil")
                     & (df["Overall"] >= 90), ["Overall"]].count())
count_overall

plt.pie(count_overall, autopct='%1.1f%%', labels = xA)
```

```python
df = pd.read_excel("pokemon_data.xlsx")
df
```

```python
lst_type1 = ["Grass", "Fire", "Electric", "Rock"]

# HP Average
lst_hp_avg = [0, 0, 0, 0]
lst_hp_avg[0] = df.loc[ df["Type 1"] == "Grass"  , "HP"].mean()
lst_hp_avg[1] = df.loc[ df["Type 1"] == "Fire"  , "HP"].mean()
lst_hp_avg[2] = df.loc[ df["Type 1"] == "Electric"  , "HP"].mean()
lst_hp_avg[3] = df.loc[ df["Type 1"] == "Rock"  , "HP"].mean()

# Attack Average
lst_atk_avg = [0, 0, 0, 0]
lst_atk_avg[0] = df.loc[ df["Type 1"] == "Grass"  , "Attack"].mean()
lst_atk_avg[1] = df.loc[ df["Type 1"] == "Fire"  , "Attack"].mean()
lst_atk_avg[2] = df.loc[ df["Type 1"] == "Electric"  , "Attack"].mean()
lst_atk_avg[3] = df.loc[ df["Type 1"] == "Rock"  , "Attack"].mean()

plt.plot(lst_type1, lst_hp_avg, color = "red")
plt.plot(lst_type1, lst_atk_avg)
```

```python
df = pd.read_excel("dienthoai.xlsx")
df
```

```python
tong_gia = []
for i in range (len(df)):
    gia = df.iloc[i,6]
    sl = df.iloc[i,7]
    tong_gia.append(gia * sl)
print(tong_gia)

lst_year_sale = []
for i in range(len(df)):
    year = df.iloc[i, 1]
    year = int(year[:4])
    lst_year_sale.append(year)

df1 = {
    "Year": lst_year_sale,
    "Price": tong_gia
}
df1 = pd.DataFrame(df1)

years = list(df1["Year"].value_counts().index)
years = sorted(years)

lst_avg = []
for y in years:
    avg = df1.loc[ df1["Year"] == y , "Price"].mean()
    lst_avg.append(avg)

plt.bar(years, lst_avg)

```

```python
df = pd.read_csv("smartphone.csv")
df
```

```python
lst_brand = list(df["brand_name"].value_counts().index)

lst_avg_price = []
for brand in lst_brand:
    rate = df.loc[ df["brand_name"] == brand   , "price"  ].mean()
    lst_avg_price.append(rate)

dic = {
    "Brand": lst_brand,
    "Price": lst_avg_price
}
df1 = pd.DataFrame(dic)
df1 = df1.sort_values(by=['Price'], ascending=False)
df1 = df1.iloc[:5]
Ox = df1["Brand"]
Oy = df1["Price"]
plt.bar(Ox, Oy)
```
