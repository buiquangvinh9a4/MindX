```python
#1
soluongbanchaynhat = df["Số lượng bán được"].max()
loaimathang = df.loc[df["Số lượng bán được"] == soluongbanchaynhat, :]
loaimathang
```

```python
#2
tong_gia = []
for i in range (len(df)):
    gia = df.iloc[i,6]
    sl = df.iloc[i,7]
    tong_gia.append(gia * sl)

print(tong_gia)

lst_year_sale = []
for i in range(len(df)):
    year = df.iloc[i, 1]
    year = int(year[:4])
    lst_year_sale.append(year)

df1 = {
"Year": lst_year_sale,
"Price": tong_gia
}
df1
df1 = pd.DataFrame(df1)
years = list(df1["Year"].value_counts().index)
years = sorted(years)
    
lst_avg = []
for y in years:
    avg = df1.loc[ df1["Year"] == y , "Price"].mean()
    lst_avg.append(avg)
    
plt.bar(years, lst_avg, width= 0.3 )
plt.plot(years, lst_avg, color = "red")
plt.scatter(years, lst_avg, color = "pink", linewidth = 0.2)
plt.title("Trung bình giá của mỗi loại sản phẩm bán ra trong từng năm", fontsize = "11")
```
