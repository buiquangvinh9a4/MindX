```python
# max 
s1 = df["Overall"].max()
print(f"Overall cao nhất: {s1}")

# min 
s2 = df["Overall"].min()
print(f"Overall thấp nhất: {s2}")

# sum 
s3 = df["Overall"].sum()
print(f"Tổng Overall : {s3}")

# count 
s4 = df["Overall"].count()
print(f"Đếm số lượng Overall : {s4}")

# mean <-> average
s5 = df["Overall"].mean()
print(f"Trung bình Overall : {s5}")

"""
    BÀI TẬP THỰC HÀNH:
    1. Hiển thị thông tin cầu thủ có số tuổi cao nhất.
    2. Hiển thị thông tin cầu thủ có Overall cao nhất.
    3. Hiển thị trung bình Overall của các cầu thủ France.
    4. Đếm số lượng cầu thủ có Overall lớn hơn 80 ở Portugal.
"""
```

```python
#1. Hiển thị thông tin cầu thủ có số tuổi cao nhất.
age_max = df["Age"].max()
result1 = df.loc[ df["Age"] == age_max ,  : ]
result1
```

```python
 #   2. Hiển thị thông tin cầu thủ có Overall cao nhất.
overall_max = df["Overall"].max()
result2 = df.loc[ df["Overall"] == overall_max , : ]
result2
```

```python
#3. Hiển thị trung bình Overall của các cầu thủ France.
result3 = df.loc[ df["National_team"] == "France"   , : ]
avg_overall = result3["Overall"].mean()
print(f"Trung bình Overall cầu thủ Pháp: {avg_overall:.4f}")
```

```python
# 4. Đếm số lượng cầu thủ có Overall lớn hơn 80 ở Portugal.
df_tmp = df.loc[ df["National_team"] == "Portugal"   , : ]
result4 = df_tmp.loc[ df_tmp["Overall"] > 80  ,  : ]

# Cách 1: Sử dụng hàm len()
print(f"Số lượng cầu thủ có Overall > 80 tại Portugal: {len(result4)}")

# Cách 2: Sử dụng hàm count()
countx = result4["Player_name"].count()
print(f"Số lượng cầu thủ có Overall > 80 tại Portugal: {countx}")
```

```python
import numpy as np 

def sigmoid(s):   
    return 1/(1 + np.exp(-s))

def logistic_sigmoid_regression(X, y, w_init, eta, tol = 1e-4, max_count = 10000):
    w = [w_init]    
    it = 0
    N = X.shape[1]
    d = X.shape[0]
    count = 0
    check_w_after = 20
    while count < max_count:
        # mix data 
        mix_id = np.random.permutation(N)
        for i in mix_id:
            xi = X[:, i].reshape(d, 1)
            yi = y[i]
            zi = sigmoid(np.dot(w[-1].T, xi))
            w_new = w[-1] + eta*(yi - zi)*xi
            count += 1
            # stopping criteria
            if count%check_w_after == 0:                
                if np.linalg.norm(w_new - w[-check_w_after]) < tol:
                    return w
            w.append(w_new)
    return w
eta = .05 
d = X.shape[0]
w_init = np.random.randn(d, 1)

w = logistic_sigmoid_regression(X, y, w_init, eta)
print(w[-1])
```
