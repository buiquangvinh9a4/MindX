```python
import time
import matplotlib.pyplot as plt

def fibo(n):
    if n <= 2:
        return 1
    return fibo(n - 1) + fibo(n - 2)

lst_time = []
value = []

for i in range(30, 37):
    start_time = time.time()
    fibo(i)
    real_time = time.time() - start_time
    lst_time.append(real_time)
    value.append(i)

plt.plot(value, lst_time)
```
