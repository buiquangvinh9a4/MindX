```python
# !!! YÊU CẦU:  Viết 1 hàm isPrimeNumber() để kiểm tra number có phải là 1 số nguyên tố?
        # Hàm có giá trị trả về TRUE or FALSE
    
    # mở rộng: khởi tạo 1 list chứa các số ngẫu nhiên 
    # yêu cầu: đếm trong list có bao nhiên số nguyên tố -> sử dụng hàm vừa viết
  

import random    # randint(): sinh ra các số nguyên t
def isPrimeNumber(number):
    if number < 2:
        return False
    else:
        for i in range(2, int(number ** 0.5) + 1):
            if number % i == 0:
                return False
        return True

lst = []
for i in range(10):
    number = random.randint(0, 100)
    lst.append(number)
    
count = 0
print(lst)

for number in lst:
    if isPrimeNumber(number) == True:
        count = count + 1

print(count)
```

```python
# Hàm nhập bốn số.
def inputNumber():
    lst = []
    for i in range(4):
        number = int(input())
        lst.append(number)
    return lst
    

# Hàm tính trung bình cộng của bốn số.
def averageNumber(lst):
    return sum(lst) / len(lst)

# Hàm trả về số lớn nhất trong bốn số.
def maxNumber(lst):
    return max(lst)

#Hàm trả về số nhỏ nhất trong bốn số.
def minNumber(lst):
    return min(lst)


lst = inputNumber()
print(averageNumber(lst))
print(maxNumber(lst))
print(minNumber(lst))


```
